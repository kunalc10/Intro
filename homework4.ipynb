```python
import pandas as pd
import numpy as np
```

```python
x = pd.read_excel('C:\\Users\\kunal\\Downloads\\gdplev.xls', header = None, skiprows =220, parse_cols = [4,6] )
x.rename(columns = {0:'Quarter', 1:'GDP'}, inplace = True)
x['diff'] = x['GDP'].diff()
def rec_converter(cell):
    if cell >= 0:
        cell = 1
        return cell
    else: 
        cell = 0
        return cell
x['rec'] = x['diff'].apply(rec_converter)
def rec_teller(cell):
    if cell == 0:
        return 'decline'
    else:
        return 'increase'
x['id'] = x['rec'].apply(rec_teller)
x.drop(67, inplace = True)
```

```python
p_id = list(x['id'])
p_quarter = list(x['Quarter'])
recession = []
st = 'decline'
st1 = 'increase'
for k1,(k2,k3) in enumerate(zip(p_id,p_quarter)):
    var = 0
    if ((k2 == st) and (p_id[k1+1] == st)):
        recession.append(k3)
        recession.append(p_quarter[k1 +1])
        if ((p_id[k1+2] == st1) and (p_id[k1+3] == st1)):
            recession.append(p_quarter[k1 +2])
            recession.append(p_quarter[k1 +3])
            break
s = set(recession)
recession = list(s)
recession.sort()
```
