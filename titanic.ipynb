```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('F:\\machine learning\\datasets\\titanic\\tt.csv')
df_test = pd.read_csv('F:\\machine learning\\datasets\\titanic\\testt.csv')
df

```

```python
df['Cabin'].fillna('No', inplace = True)
df_test['Cabin'].fillna('No', inplace = True)
def cabin(cell):
    if cell =="No":
        return cell
    else:
        return cell[0]
df['Cabin'] = df['Cabin'].apply(cabin)
df_test['Cabin'] = df_test['Cabin'].apply(cabin)
#df_test['Cabin']
#df[['Cabin', 'Survived']].groupby(by = 'Cabin').mean()
df_test['Cabin'].value_counts()
```

```python
xxx = {'A' : 1, 'B': 2, 'C': 3, 'D':4, 'E': 5, 'F': 6, 'G':7,'T':8, 'No' : 9}
df.set_index('Cabin', inplace = True)
df.rename(index = xxx, inplace = True)
df.reset_index(inplace = True)
df_test.set_index('Cabin', inplace = True)
df_test.rename(index = xxx, inplace = True)
df_test.reset_index(inplace = True)
```

```python
def title(cell):
    cell = (cell['Name'])
    cell = cell.split(',')[1]
    cell = cell.split('.')[0]
    cell = cell.lstrip()
    return cell
df['Title'] = df.apply(title, axis = 1)
df_test['Title'] = df_test.apply(title, axis = 1)
df[['Title', 'Survived']].groupby('Title').mean()
df['Title'].value_counts()
df_test['Title'].value_counts()

    
    
```

```python
def change(cell):
    if (cell == 'the Countess') | (cell == 'Sir') | (cell == 'Jonkeer') | (cell == 'Don') | (cell == 'Capt') | (cell == 'Lady') | (cell =='Major') | (cell == 'Col') | (cell == 'Dona'):
        return 1
    elif (cell == 'Mrs') | (cell == 'Mme'):
        return 2
    elif cell =='Rev':
        return 3
    elif cell == 'Dr':
        return 4
    elif (cell == 'Miss') | (cell == 'Mlle') | (cell == 'Ms'):
        return 5
    elif (cell == 'Mr'):
        return 6
    elif (cell == 'Master'):
        return 7
df['New_title'] = df['Title'].apply(change)
df_test['New_title'] = df_test['Title'].apply(change)
df_test
df[['New_title', 'Survived']].groupby('New_title').mean()
df['New_title'].corr(df['Survived'])
df_test[df_test['New_title'].isnull()]


    
```

```python
from sklearn.ensemble import RandomForestClassifier
features = pf[['New_feature', 'Embarked','Fare','Family','New_feature2','New_title','Age','Cabin','Isalone']].values
target = pf['Survived'].values
from sklearn.cross_validation import train_test_split
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size = 0.25)
clf = RandomForestClassifier(n_estimators = 100)
a = []
for i in range(10):
    clf = clf.fit(features_train, target_train)
    predicted_train = clf.predict(features_train)
    predicted_test = clf.predict(features_test)
    a.append(predicted_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(target_train, predicted_train))
print(accuracy_score(target_test, predicted_test))
```

```python
b = np.array(a)
from scipy import stats
pre_loop = stats.mode(b)
pre_loop = pre_loop[0]
pre_loop = pre_loop.T
print(accuracy_score(target_test, pre_loop))
```
