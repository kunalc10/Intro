```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('F:\\machine learning\\datasets\\comp1\\train_xup5Mf8\\train.csv')
train.head()
test = pd.read_csv('F:\\machine learning\\datasets\\comp1\\test_6lvBXoI\\test.csv')
print(test.shape)
print(train.shape)
train.columns
```

```python
train1 = train.groupby(by = 'Stock_ID')
a = []
for key, grouped in train1:
    a.append(grouped)
print(len(a))
```

```python
test1 = test.groupby(by = 'Stock_ID')
b = []
for key, grouped in test1:
    b.append(grouped)
print(len(b))
#b[0]
```

```python
train_stock1 = a[0]
test_stock1 = b[0]
test_stock1.isnull().any()
```

```python

from sklearn.ensemble import RandomForestClassifier
features = train_stock1[['Volume', 'Three_Day_Moving_Average',
       'Five_Day_Moving_Average', 'Ten_Day_Moving_Average',
       'Twenty_Day_Moving_Average', 'True_Range', 'Average_True_Range',
       'Positive_Directional_Movement', 'Negative_Directional_Movement']]
target = train_stock1['Outcome']
clf = RandomForestClassifier()
clf = clf.fit(features, target)
test_features = test_stock1[['Volume', 'Three_Day_Moving_Average',
       'Five_Day_Moving_Average', 'Ten_Day_Moving_Average',
       'Twenty_Day_Moving_Average', 'True_Range', 'Average_True_Range',
       'Positive_Directional_Movement', 'Negative_Directional_Movement']]
test_stock1['Outcome'] = clf.predict(test_features)
test_stock1
```
